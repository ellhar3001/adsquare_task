```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import datetime
```

```
data = pd.read_csv('/content/final_data.csv')
data.drop('Unnamed: 0',axis = 1, inplace = True)
```

```
data['DateTime'] = pd.to_datetime(data['date'])
data
```

```
data['stores'].unique()
```

```
McDonalds = data[data['stores']=="McDonald's"]
Burger_King = data[data['stores']=="Burger King"]
Subway = data[data['stores']=="Subway"]
Aldi = data[data['stores']=="Aldi"]
```

```
plt.figure(figsize=(20,8))

g = sns.lineplot(data=McDonalds, x="DateTime", y="unique visitors",hue = 'store_id')
g.set_title('McDonalds Unique Visitors')
g.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

```
plt.figure(figsize=(20,8))

g = sns.lineplot(data=Burger_King, x="DateTime", y="unique visitors",hue = 'store_id')
g.set_title('Burger King Unique Visitors')
g.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```

```
plt.figure(figsize=(20,8))

g = sns.lineplot(data=Aldi, x="DateTime", y="unique visitors",hue = 'store_id')
g.set_title('Aldi Unique Visitors')
g.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)
```
